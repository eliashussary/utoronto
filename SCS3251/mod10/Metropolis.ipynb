```python
import numpy as np
import matplotlib.pyplot as plt
import matplotlib.mlab as mlab
```

```python
def q(x, y):
    g1 = mlab.bivariate_normal(x, y, 1.0, 1.0, -1, -1, -0.8)
    g2 = mlab.bivariate_normal(x, y, 1.5, 0.8, 1, 2, 0.6)
    return 0.6*g1 + 28.4*g2/(0.6+28.4)
```

```python
# Metropilis 
N = 100000
s = 50
r= np.zeros(2)
p = q(r[0], r[1])

samples = []

for i in range(N):
    rn = r + np.random.normal(size=2)
    pn = q(rn[0], rn[1])
    if pn >= p:
        p = pn
        r = rn
    else:
        u = np.random.rand()
        if u < pn/p:
            p = pn
            r = rn
    if i % s ==0:
        samples.append(r)
        
```

```python
samples = np.array(samples)
plt.scatter(samples[:, 0], samples[:, 1], alpha=0.5, s=1)
plt.show()
# Plot target
dx= 0.01
x = np.arange(np.min(samples), np.max(samples), dx)
y = np.arange(np.min(samples), np.max(samples), dx)
X, Y = np.meshgrid(x, y)
Z = q(X, Y)
CS = plt.contour(X, Y, Z, 5)
#plt.clabel(CS, inline=1, fontsize=20)
```
