```python
def plot_and_describe(x, bins):
    mean = np.mean(x)
    var = np.var(x)
    std = np.std(x)
    skewness = stats.stats.skew(x)
    kurtosis = stats.stats.kurtosis(x)

    df = pd.Series([mean, var, std, skewness, kurtosis],
            index=['mean', 'variance', 'std.', 'skewness', 'kurtosis'])
    df = pd.DataFrame(df, columns=['value'])
    display(df)
    #plot
    plt.figure(figsize=(10,5))
    ax = plt.subplot(121)
    ax.plot(x)
    ax.set_xlabel("x")
    
    ax = plt.subplot(122)
    ax.hist(x, bins=bins)
    ax.set_ylabel("count")
    ax.set_xlabel("hist (x)")
    plt.tight_layout()
    plt.show()
```

```python
plot_and_describe(x,10)
```

### Plot low, high std

```python
x_axis = np.arange(-20, 20, 0.001)
mu=0.0
std=8.0

std2=1.0

plt.plot(x_axis, 3.0*stats.norm.pdf(x_axis,mu,std), label='high std. - wide')
plt.plot(x_axis, stats.norm.pdf(x_axis,mu,std2), label='low std. - thin')

plt.ylabel('pdf')
plt.xlabel('x')
plt.legend(loc='best')
plt.grid(False)
plt.show()
```

### Uniform Distribution

```python
r=stats.uniform.rvs(size=1000)
ax = plt.hist(r, bins=25, normed=True, histtype='stepfilled', alpha=0.2, label='uniform')
plt.legend(loc='best', frameon=False)
plt.plot(ax[1], np.ones_like(ax[1]), linewidth=2)
plt.show()
```

```python
plot_and_describe(stats.uniform.rvs(size=1000), 20)
```

```python
plot_and_describe(stats.uniform.rvs(size=100000), 20)
```

```python
u = stats.uniform()
```

```python
x = np.linspace(u.ppf(0.01), u.ppf(0.99), 100)

plt.figure(figsize=(10,15))
ax = plt.subplot(311)
plt.plot(u.rvs(size=1000),'r-', lw=5, alpha=0.6, label='sample')
plt.legend(loc='best')

ax = plt.subplot(312)
plt.plot(x, u.pdf(x),'r-', lw=5, alpha=0.6, label='uniform pdf')
plt.legend(loc='best')

ax = plt.subplot(313)
plt.plot(x, u.cdf(x),'r-', lw=5, alpha=0.6, label='uniform cdf')
plt.legend(loc='best')

plt.tight_layout()
plt.show()
```

## normal distribution

```python
# Plot between -10 and 10 with .001 steps.
@interact(mu=(0.0,10.0,1.0), std=(1.0,10.0,1.0))
def plot_normal(mu=0.0, std=2.0):
    x_axis = np.arange(-10, 10, 0.001)
    plt.plot(x_axis, stats.norm.pdf(x_axis,mu,std))
    plt.show()
```
