```python
import statsmodels.api as sm
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt
```

Load and confirm there are 6 datasets loaded.

```python
sets = pd.read_excel("Module6_Exercise.xlsx", sheet_name=None)
print("Num of datasets:", len(sets))
```

---
#### Run The Linear Regression
Prepare our models by iterating over each set, and running our linear regression fit. We'll store each model as a dictionary within a list, while exposing x, y sets.

```python
models = []

for set_name in sets:
    dataset = sets[set_name]
    y, X = dataset['y'], dataset['x']
    sm_model = sm.OLS(y, sm.add_constant(X)).fit()
    residuals = y - sm_model.predict(sm.add_constant(X))
    models.append({
        'name': set_name,
        'model': sm_model,
        'set': sets[set_name],
        'residuals': residuals
    })
```

Define a helper function to plot a model.

```python
def plotRegression(model):
    intercept, slope = model['model'].params
    x = model['set'].x
    y = model['set'].y
    plt.title(model['name'] + ' - Linear Regression')
    plt.plot(x, y, 'o', label='data_point', markersize=5)
    plt.plot(x, intercept + slope * x, 'r-', label='fitted_line')
    plt.legend()
```

Plot each model in it's own subplot.

```python
plt.figure(figsize=(24,12))

for i in range(len(models)):   
    plt.subplot(231+ i)
    plotRegression(models[i])
```

#### Analysis of Regression
Do a graphical analysis: a. plot histogram of residuals; b. plot residuals vs. predictor.

```python
def plotResiduals(model):
    
    #destructure
    set, residuals, name, sm_model = model['set'], model['residuals'], model['name'], model['model']
    y, X = set['y'], set['x']
        
    #plot
    plt.figure(figsize=(24,6))
    
    #regression
    plt.subplot(131)
    plotRegression(model)
    
    # histogram
    plt.subplot(132)
    plt.hist(residuals, bins=10)
    plt.title(name + ' - Histogram')
    
    # residuals v predictor
    plt.subplot(133)
    plt.plot(X, residuals, 'o')
    plt.axhline(y=0, color='r', linestyle='-')
    plt.title(name + ' - Residuals vs Predictor')
    
    print(sm_model.summary())
```
