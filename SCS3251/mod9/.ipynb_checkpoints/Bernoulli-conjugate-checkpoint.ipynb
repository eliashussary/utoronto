```python
%matplotlib inline

from IPython.core.pylabtools import figsize
import numpy as np
import pandas as pd
from matplotlib import pyplot as plt
import seaborn.apionly as sns
figsize(11,9)
```

```python
import scipy.stats as stats

dist = stats.beta
n_trials = [0,1,2,3,4,5,8,15,50,500]
data = stats.bernoulli.rvs(0.5,size=n_trials[-1])
x = np.linspace(0, 1, 100)
```

```python
# Using Bernoulli conjugate prior.

for k, N in enumerate(n_trials):
    sx=plt.subplot(len(n_trials)/2, 2, k+1)
    plt.xlabel("$p$, probability of heads") if k in [0, len(n_trials)-1] else None
    plt.setp(sx.get_yticklabels(), visible=False)
    heads = data[:N].sum()
    y = dist.pdf(x, 1 + heads, 1+ N - heads)
    plt.plot(x, y, label="observe %d tosses,\n %d heads" % (N, heads))
    plt.fill_between(x, 0, y, color="#348ABD", alpha=0.4)
    plt.vlines(0.5, 0, 4, color = "r", linestyles="--", lw=2)
    leg = plt.legend()
    leg.get_frame().set_alpha(0.4)
    plt.autoscale(tight=True)
    
plt.suptitle("Bayesian updating of posterior probabilities", y=1.02, fontsize=14)
plt.tight_layout
```

```python
figsize(12.5, 3.5)
message_data = np.loadtxt("data/txtdata.csv")
n_message_data = len(message_data)
plt.bar(np.arange(n_message_data), message_data, color="#348ABD")
plt.xlabel("Time (days)")
plt.ylabel("count of messages received")
plt.title("Did the users's texting habits change over time?")
plt.xlim(0, n_message_data)
```

```python
import pymc3 as pm
import theano.tensor as tt

with pm.Model() as model:
    alpha = 1.0/(message_data.mean())
    lambda_1 = pm.Exponential("lambda_1", alpha)
    lambda_2 = pm.Exponential("lambda_2", alpha)
    
    tau = pm.DiscreteUniform("tau", lower=0, upper=n_message_data - 1)
    
```

```python
with model:
    idx = np.arange(n_message_data)
    lambda_ = pm.math.switch(tau >= idx, lambda_1, lambda_2)
```

```python
with model:
    observation = pm.Poisson("obs", lambda_, observed = message_data)
```

```python
with model:
    step = pm.Metropolis()
    trace = pm.sample(10000, tune = 5000, step=step)
```

```python
lambda_1_samples = trace['lambda_1']
lambda_2_samples = trace['lambda_2']
tau_samples = trace['tau']
```

```python
figsize(12.5, 10)
ax=plt.subplot(311)
ax.set_autoscaley_on(False)

plt.hist(lambda_1_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_1$", color="#A60628", normed = True)
plt.legend(loc="upper left")
plt.title(r"""Posterior distributions of the variables
    $\lambda_1,\;\lambda_2,\;\tau$""")
plt.xlim([15, 30])
plt.xlabel("$\lambda_1$ value")

ax=plt.subplot(312)
ax.set_autoscaley_on(False)

plt.hist(lambda_2_samples, histtype='stepfilled', bins=30, alpha=0.85,
         label="posterior of $\lambda_2$", color="#7A68A6", normed = True)
plt.legend(loc="upper left")
plt.xlim([15, 30])
plt.xlabel("$\lambda_2$ value")

ax=plt.subplot(313)
w=1.0 / tau_samples.shape[0] * np.ones_like(tau_samples)
plt.hist(tau_samples, bins=n_message_data, alpha=1,
         label=r"posterior of $\tau$", color="#467821", weights=w, rwidth=2.)
plt.xticks(np.arange(n_message_data))

plt.legend(loc="upper left")
plt.ylim([0, .75])
plt.xlim([15, len(message_data)-20])
plt.xlabel(r"$\tau$ (in days)")
plt.ylabel("probability")
```

```python
figsize(12.5, 5)
N_tau=tau_samples.shape[0]
expected_texts_per_day = np.zeros(n_message_data)
for day in range(0, n_message_data):
    # ix is a bool index of all tau samples corresponding to
    # the switchpoint occuring prior to value of 'day'
    ix = day < tau_samples
    expected_texts_per_day[day]=(lambda_1_samples[ix].sum()+lambda_2_samples[~ix].sum())/N_tau
    
plt.plot(range(n_message_data), expected_texts_per_day, lw=4, color="#E24A33",
        label="expected number of text-messages received")
plt.xlim(0, n_message_data)
plt.xlabel("Day")
plt.ylabel("Expected # of text messages")
plt.title("Expected number of text-messages received")
plt.ylim(0, 60)
plt.bar(np.arange(len(message_data)), message_data,color="#348ABD", alpha=0.65,
       label="observed texts per day")
plt.legend(loc="upper left")
```
